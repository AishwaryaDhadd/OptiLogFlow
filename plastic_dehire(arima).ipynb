```python
import ThymeBoost as tb
import sklearn as sk
```

```python
#outlier treatment
from ThymeBoost import ThymeBoost as tb
boosted_model = tb.ThymeBoost()
```

```python
output = boosted_model.detect_outliers(p1_new['QUANTITY'],
                                       trend_estimator='linear',
                                       seasonal_estimator='fourier',
                                       seasonal_period=15,
                                       global_cost='maicc',
                                       fit_type='global')
```

```python
boosted_model.plot_results(output)
boosted_model.plot_components(output)
```

```python
#Handling Outliers Using the Regularize Argument


output_r = boosted_model.fit(p1_new['QUANTITY'],
                           trend_estimator='linear',
                           seasonal_estimator='fourier',
                           seasonal_period=15,
                           global_cost='maicc',
                           fit_type='global',
                           seasonality_weights='regularize')
```

```python
boosted_model.plot_results(output)

boosted_model.plot_components(output)
```

```python
rmse_r = np.sqrt(np.mean((np.array(output_r['y'])-np.array(output_r['yhat']))**2))
rmse_r
```

```python
###weight
weights = np.invert(output['outliers'].values) * 1
```

```python
output_w = boosted_model.fit(p1_new['QUANTITY'],
                           trend_estimator='linear',
                           seasonal_estimator='fourier',
                           seasonal_period=15,
                           global_cost='maicc',
                           fit_type='global',
                           seasonality_weights=weights)
```

```python
boosted_model.plot_results(output_w)
boosted_model.plot_components(output_w)
```

```python
rmse_w = np.sqrt(np.mean((np.array(output_w['y'])-np.array(output_w['yhat']))**2))
rmse_w
```

```python
import statsmodels.api as sm
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
#Stationary Check
X = p1_new["QUANTITY"].values
```

```python
result = adfuller(X)
```

```python
print('ADF Statistic: %f' % result[0])
```

```python
print('p-value: %f' % result[1])
```

```python
print('Critical Values:')
for key, value in result[4].items(): print('\t%s: %.3f' % (key, value))
```

```python
if result[0] < result[4]["5%"]: 
    print ("Reject Ho - Time Series is Stationary")
else:
    print("Failed to Reject Ho - Time Series is Non-Stationary")
```

```python
arima_model = auto_arima(train.QUANTITY,start_p=0, start_q=0, max_p=12, max_q=12,start_P = 0,
                          m=1,d = None,
                         seasonal=False, error_action='warn',
                         trace=True, stepwise=True)
```

```python
arima_model.summary()
```

```python
prediction = pd.DataFrame(arima_model.predict(n_periods =191))
prediction.columns = ['QUANTITY']
prediction
```

```python
plt.plot(prediction,label="Predicted")
```

```python
plt.plot(test,label="Test")
```

```python
plt.figure(figsize=(8,5))
plt.plot(train,label="Training")
plt.plot(test,label="Test")
plt.plot(prediction,label="Predicted")
plt.show()
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error

rmse_best = sqrt(mean_squared_error(test.QUANTITY, prediction))
print('Test RMSE: %.3f' % rmse_best)
```
